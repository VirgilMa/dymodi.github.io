```python
# Here we to massive test to check the hypothesis
# Yi Ding
# 07/12/18

import numpy as np

# First we test a small net with two shops A and B
# Also not S is the source node
max_iter = 1000
for i in range(0,max_iter):
    e_SA = np.ceil(10 * np.random.rand(1))
    e_SB = np.ceil(10 * np.random.rand(1))
    e_AB = np.ceil(10 * np.random.rand(1))
    if e_SA > e_SB + e_AB or e_SB > e_SA + e_AB or e_AB > e_SA + e_SB:
        # Must be triangle
        continue
    w_A = np.ceil(20 * np.random.rand(1))
    w_B = np.ceil(20 * np.random.rand(1))
    print(i)
    print('e_SA:',e_SA)
    print('e_SB:',e_SB)
    print('e_AB:',e_AB)
    print('w_A:',w_A)
    print('w_B:',w_B)
    # Calculate route S->A->B
    t_leaveA = max([e_SA,w_A])
    time_SAB = t_leaveA + max([t_leaveA+e_AB, w_B])
    print('time_SAB:',time_SAB)
    
    # Calculate route S->B->A
    t_leaveB = max([e_SB,w_B])
    time_SBA = t_leaveB + max([t_leaveB+e_AB, w_A])
    print('time_SBA:',time_SBA)
    
    # The time greedy?
    if t_leaveA < t_leaveB:
        if time_SAB > time_SBA:
            raise Exception('Greedy not best?')

```
